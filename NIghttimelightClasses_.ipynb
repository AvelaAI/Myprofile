```python
import pandas as pd
from datetime import datetime, timedelta
import plotly.express as px

# pd.set_option('future.no_silent_downcasting', True)

# df = pd.read_csv("CPT.csv")
# px.bar(df, x="Urban",y="system:Year")

import pandas as pd
import matplotlib.pyplot as plt

# Define the CSV data
data = {
    'Year': ['2019', '2020', '2021'],
    'Urban': ['122,317.116', '116,190.431', '107,296.625'],
    'Recreational': ['18,216.176', '18,620.501', '17,731.968'],
    'Water': [535.44, 512.04, 464.501],
    'Barren': [535.44, 512.04, 464.501]
}

# Convert the data into a DataFrame
df = pd.DataFrame(data)

# Convert the 'Urban' and 'Recreational' columns to numeric values
df['Urban'] = df['Urban'].str.replace(',', '').astype(float)
df['Recreational'] = df['Recreational'].str.replace(',', '').astype(float)

# Set up the bar plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plotting the data
df.set_index('Year')[['Urban', 'Recreational', 'Water', 'Barren']].plot(kind='bar', ax=ax)

# Adding titles and labels
plt.title('Night Lights Time Series for Cape Town Over Years', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Values (in thousands)', fontsize=12)
plt.xticks(rotation=0, ha='right')  # Rotating the x-axis labels for readability

# Display the plot
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))

# Plotting the data as a line graph
plt.plot(df['Year'], df['Urban'], label='Urban', marker='o')
plt.plot(df['Year'], df['Recreational'], label='Recreational', marker='o')
plt.plot(df['Year'], df['Water'], label='Water', marker='o')
plt.plot(df['Year'], df['Barren'], label='Barren', marker='o')

# Adding titles and labels
plt.title('Class Systems Over Years (Line Graph)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Values (in thousands)', fontsize=12)
plt.xticks(rotation=45, ha='right')  # Rotating the x-axis labels for readability

# Displaying legend
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
```
