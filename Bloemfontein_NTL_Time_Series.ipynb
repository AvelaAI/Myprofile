```python
import pandas as pd
import matplotlib.pyplot as plt

# Create the DataFrame with the data
data = {
    'system:time_start': ['2018', '2019', '2020', '2021'],
    'Built up': [11282.427, 11015.1069, 10020, 10003.4298],
    'Impervious Surfaces': [4835.325, 4720.7601, 3824, 4287.1842],
    'Recreational': [5507.9976, 5294.2224, 6076.3512, 6236.044]
}

df = pd.DataFrame(data)


# Set the system:time_start column as index
df.set_index('system:time_start', inplace=True)

# Create the bar plot
df.plot(kind='bar', figsize=(10, 6))

# Add title and labels
plt.title('Yearly Land Cover')
plt.xlabel('Year')
plt.ylabel('Values')
plt.xticks(rotation=0)

# Show the plot
plt.tight_layout()
plt.show()


# Create the line plot
df.plot(kind='line', figsize=(10, 6), marker='o')

# Add title and labels
plt.title('Nighttime Lights Time Series for Bleomfontein')
plt.xlabel('Year')
plt.ylabel('Sum of Nighttime Lights')
plt.xticks(rotation=0)
plt.legend(title='Categories')

# Show the plot
plt.tight_layout()
plt.show()
```
