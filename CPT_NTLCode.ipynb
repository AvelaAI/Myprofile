```python
import matplotlib.pyplot as plt
import pandas as pd

# Data for the time series
# Data for the time series
data = {
    'system:time_start': ['2018', '2019', '2020', '2021'],
    'Built up': [75212.4562, 77851.7684, 72428.6934, 73166.9274],
    'Impervious Surfaces': [32233.9098, 33365.0436, 31040.8686, 31357.2546],
    'Recreational': [19236.7292, 19867.2868, 20906.96, 21364.8267]
}

# Display the data
import pandas as pd

# Create DataFrame
df = pd.DataFrame(data)

# Convert the 'system:time_start' to datetime for sorting (if needed)
df['system:time_start'] = pd.to_datetime(df['system:time_start'], format='%Y')

# Sort by 'system:time_start'
df_sorted = df.sort_values('system:time_start')

# Show the sorted DataFrame
print(df_sorted)

# df =pd.read_csv("BFN.csv")
df = pd.DataFrame(data)


# Set the system:time_start column as index
df.set_index('system:time_start', inplace=True)

# Create the bar plot
df.plot(kind='bar', figsize=(10, 6))

# Add title and labels
plt.title('Land Cover Classes')
plt.xlabel('Year')
plt.ylabel('Values')
plt.xticks(rotation=0)

# Show the plot
plt.tight_layout()
plt.show()


# Create the line plot
df.plot(kind='line', figsize=(10, 6), marker='o')

# Add title and labels
plt.title('Nighttime Lights Time Series for Pretoria')
plt.xlabel('Year')
plt.ylabel('Sum of Nighttime Lights (W/sr/m)')
plt.xticks(rotation=0)
# plt.legend(title='Categories')

# Show the plot
plt.tight_layout()
plt.show()
```
