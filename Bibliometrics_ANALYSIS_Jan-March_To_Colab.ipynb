```
!pip install wordcloud matplotlib pillow numpy requests

import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from PIL import Image
import requests
from io import BytesIO
from matplotlib.colors import LinearSegmentedColormap

# -------------------------------------
# CONFIGURATION — EDIT THESE!
# -------------------------------------

shape_choice = "leaf"     # options: "circle", "leaf", "satellite"
color_theme = "greens"    # options: "greens", "water", "heat"
font_path = None          # example: "Roboto-Regular.ttf" (upload first)

# -------------------------------------
# WORD FREQUENCIES
# -------------------------------------
data = {
    "Remote sensing":430, "Vegetation":125, "Land use":117, "grassland":105,
    "agriculture":79, "land cover":78, "ndvi":57, "satellite imagery":56,
    "biomass":55, "forestry":55, "modis":54, "ecosystems":52, "water quality":52,
    "gis":50, "climate change":45, "land use change":45, "ecosystem":44,
    "biodiversity":41, "geographic information systems":40, "rangeland":38,
    "landsat":37, "remote-sensing":37, "classification":36, "nitrogen":35,
    "ecology":34, "forest":34, "image analysis":34, "landscape":34,
    "management":34, "environmental monitoring":32, "dynamics":31, "pasture":31,
    "urbanization":30, "degradation":29, "ecosystem services":29, "monitoring":29,
    "vegetation index":29, "index":27, "soil":27, "climate-change":26,
    "productivity":26, "satellite data":26, "environmental protection":25,
    "patterns":25, "quality":25, "sustainable development":25, "cover":24,
    "habitat quality":24
}

# -------------------------------------
# SHAPE MASKS
# -------------------------------------
def load_circle_mask(size=1500):
    x, y = np.ogrid[:size, :size]
    center = size // 2
    radius = size // 2 - 10
    mask = (x - center)**2 + (y - center)**2 <= radius**2
    return (mask.astype(int) * 255)

def load_mask_from_url(url):
    response = requests.get(url)
    img = Image.open(BytesIO(response.content)).convert("L")
    return np.array(img)

if shape_choice == "circle":
    mask = load_circle_mask()

elif shape_choice == "leaf":
    mask_url = "https://raw.githubusercontent.com/ultralytics/yolov5/master/data/images/leaf-mask.png"
    mask = load_mask_from_url(mask_url)

elif shape_choice == "satellite":
    mask_url = "https://raw.githubusercontent.com/thorwhalen/wordcloud/master/examples/mask_satellite.png"
    mask = load_mask_from_url(mask_url)

# -------------------------------------
# COLOR THEMES
# -------------------------------------
if color_theme == "greens":
    colors = LinearSegmentedColormap.from_list(
        "greens",
        ["#0b3d02", "#167a12", "#4caf50", "#8bc34a", "#cddc39"]
    )

elif color_theme == "water":
    colors = LinearSegmentedColormap.from_list(
        "water",
        ["#001f3f", "#005b96", "#00a8e8", "#b3ecff"]
    )

elif color_theme == "heat":
    colors = LinearSegmentedColormap.from_list(
        "heat",
        ["#4c0000", "#b30000", "#ff6600", "#ffcc00"]
    )

# -------------------------------------
# GENERATE WORD CLOUD
# -------------------------------------
wc = WordCloud(
    width=2000,
    height=2000,
    background_color="white",
    mask=mask,
    colormap=colors,
    font_path=font_path,  # custom font
    prefer_horizontal=0.9,
    max_words=200
).generate_from_frequencies(data)

# -------------------------------------
# DISPLAY
# -------------------------------------
plt.figure(figsize=(14, 14))
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()

wc.to_file("custom_wordcloud.png")
print("Saved as custom_wordcloud.png")
```
